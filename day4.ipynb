```julia
A = rand(3,3) * 100
A = round.(A)
```

```julia
# Transpose 

A'

A .*A'

transpose(A)  # non-conjugate transpose
A'            # conjugate transpose (for complex numbers)
```

```julia
A = [1 5; 3 7]
mask = A .> 4          # Matrix of true/false
A[mask]                # Extract elements > 4 → [5, 7]
```

```julia
using LinearAlgebra
det(A)  # -2.0
inv(A)  # returns the inverse of A
eigvals(A)    # Only eigenvalues
eigvecs(A)    # Only eigenvectors
eigen(A)      # returns both
rank(A)
nullspace(A)  # returns basis of nullspace
```
